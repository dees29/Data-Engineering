Create Bucket

```python
client = boto3.client('s3')
```

Read Parquet File

```python
import io
```

```python
filename='pratice.parquet'
bucketName='deepupillai972020'

buffer = io.BytesIO()
client = boto3.resource('s3')

object = client.Object(bucketName, filename)
object.download_fileobj(buffer)

df = pd.read_parquet(buffer)
df
```

Upload Pandas DataFrame to S3

```python
from io import StringIO
```

```python
filename = 'log_data.csv'
bucketName = 'deepupillai972020'

csv_buffer = StringIO()
df.to_csv(csv_buffer)

client = boto3.client('s3')

response = client.put_object(
    ACL='private',
    Body=csv_buffer.getvalue(),
    Bucket=bucketName,
    Key=filename
)
```
